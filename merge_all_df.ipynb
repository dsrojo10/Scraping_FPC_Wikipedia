```python
import os

# Listar todos los archivos en la carpeta 'csv'
csv_files = os.listdir('csv')

print(len(csv_files))
```

```python
import pandas as pd

# Lista para almacenar los dataframes
dataframes = []

# Leer cada archivo CSV y agregarlo a la lista de dataframes
for file in csv_files:
    df = pd.read_csv(os.path.join('csv', file))
    dataframes.append(df)

# Concatenar todos los dataframes en uno solo
combined_df = pd.concat(dataframes, ignore_index=True)

print(combined_df)
```

```python
# Filtrar los dataframes correspondientes a los archivos especificados
df_2007 = next(df for df, file in zip(dataframes, csv_files) if file == 'Torneo_Finalización_2007_(Colombia)_legacy.csv')
df_2004 = next(df for df, file in zip(dataframes, csv_files) if file == 'Torneo_Finalización_2004_(Colombia)_legacy.csv')

# Concatenar los dataframes
concatenated_df = pd.concat([df_2007, df_2004], ignore_index=True)

print(concatenated_df)
# Ordenar el dataframe concatenado por año y luego por fecha en orden ascendente
concatenated_df = concatenated_df.sort_values(by=['Año', 'Fecha'], ascending=True)

print(concatenated_df)
```
