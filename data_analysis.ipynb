```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Simulando el DataFrame
# Asegúrate de que `df` ya está cargado correctamente

# Limpieza del campo `Resultado`
def limpiar_resultado(resultado):
    """Extrae los números del resultado y devuelve los goles locales y visitantes."""
    import re
    match = re.search(r'(\d+)\s*-\s*(\d+)', resultado)
    if match:
        return int(match.group(1)), int(match.group(2))
    return None, None

# Aplicar la limpieza
# Crear columnas de goles locales y visitantes
df['Goles_Local'], df['Goles_Visitante'] = zip(*df['Resultado'].apply(limpiar_resultado))

# Determinar el resultado del partido
def determinar_resultado(row):
    if row['Goles_Local'] > row['Goles_Visitante']:
        return 'Local'
    elif row['Goles_Local'] < row['Goles_Visitante']:
        return 'Visitante'
    else:
        return 'Empate'

df['Resultado_Partido'] = df.apply(determinar_resultado, axis=1)

# Análisis generales
# 1. Frecuencia de resultados
frecuencia_resultados = df['Resultado'].value_counts()

# 2. Promedio de goles
# Filtrar filas válidas (sin NaN en goles)
df_validos = df.dropna(subset=['Goles_Local', 'Goles_Visitante'])

# Calcular la suma total de goles y el número de partidos
total_goles_locales = df_validos['Goles_Local'].sum()
total_goles_visitantes = df_validos['Goles_Visitante'].sum()
numero_partidos = len(df_validos)

# Calcular promedios
promedio_goles = {
    'Total': (total_goles_locales + total_goles_visitantes) / numero_partidos,
    'Local': total_goles_locales / numero_partidos,
    'Visitante': total_goles_visitantes / numero_partidos
}

# 3. Distribución de resultados
distribucion_resultados = df['Resultado_Partido'].value_counts(normalize=True) * 100

# Tendencias por Año o Torneo
# Promedio de goles por año
goles_por_ano = df_validos.groupby('Año')[['Goles_Local', 'Goles_Visitante']].mean()
goles_por_ano['Total'] = goles_por_ano['Goles_Local'] + goles_por_ano['Goles_Visitante']

# Distribución de empates por año
empates_por_ano = df[df['Resultado_Partido'] == 'Empate'].groupby('Año').size()

# Visualizaciones adicionales
plt.figure(figsize=(16, 12))

# Tendencia de promedio de goles por año
plt.subplot(3, 2, 1)
goles_por_ano[['Goles_Local', 'Goles_Visitante', 'Total']].plot(ax=plt.gca(), marker='o')
plt.title('Promedio de Goles por Año')
plt.ylabel('Promedio de Goles')
plt.xlabel('Año')
plt.grid(True)

# Tendencia de empates por año
plt.subplot(3, 2, 2)
empates_por_ano.plot(kind='bar', color='orange', edgecolor='black', ax=plt.gca())
plt.title('Cantidad de Empates por Año')
plt.ylabel('Cantidad de Empates')
plt.xlabel('Año')

# Equipos más ganadores y perdedores
victorias = df[df['Resultado_Partido'] == 'Local'].groupby('Local').size() + df[df['Resultado_Partido'] == 'Visitante'].groupby('Visitante').size()
empates = df[df['Resultado_Partido'] == 'Empate'].groupby('Local').size() + df[df['Resultado_Partido'] == 'Empate'].groupby('Visitante').size()
derrotas = df[df['Resultado_Partido'] == 'Local'].groupby('Visitante').size() + df[df['Resultado_Partido'] == 'Visitante'].groupby('Local').size()

ranking_equipos = pd.DataFrame({
    'Victorias': victorias,
    'Empates': empates,
    'Derrotas': derrotas
}).fillna(0).astype(int)

ranking_equipos['Total_Partidos'] = ranking_equipos.sum(axis=1)
ranking_equipos = ranking_equipos.sort_values(by='Victorias', ascending=False)

# Gráfico de equipos más ganadores
plt.subplot(3, 2, 3)
ranking_equipos['Victorias'].head(10).plot(kind='bar', color='green', edgecolor='black')
plt.title('Top 10 Equipos con Más Victorias')
plt.ylabel('Cantidad de Victorias')
plt.xlabel('Equipo')

# Goleadas y derrotas abultadas
# Partidos con más de 4 goles
df['Total_Goles'] = df['Goles_Local'] + df['Goles_Visitante']
partidos_goleadas = df[df['Total_Goles'] >= 5]

# Mayor cantidad de goles en un partido
max_goles_partido = df['Total_Goles'].max()
partidos_max_goles = df[df['Total_Goles'] == max_goles_partido]

# Gráfico de partidos con más de 4 goles
plt.subplot(3, 2, 4)
partidos_goleadas['Total_Goles'].value_counts().sort_index().plot(kind='bar', color='purple', edgecolor='black')
plt.title('Distribución de Partidos con Más de 4 Goles')
plt.ylabel('Cantidad de Partidos')
plt.xlabel('Total de Goles')

plt.tight_layout()
plt.show()

# Resultados principales
print("\nPromedio de goles por partido:")
print(promedio_goles)

print("\nDistribución de resultados:")
print(distribucion_resultados)

print("\nRanking de equipos (Top 10):")
print(ranking_equipos.head(10))

print("\nPartidos con más goles:")
print(partidos_max_goles)
```
